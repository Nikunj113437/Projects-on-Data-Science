```python
#Liking top 30 posts
import time
post_btn=driver.find_element(By.CLASS_NAME,'_aa_w')
post_btn.click()

time.sleep(2)
count=1

post=driver.find_element(By.XPATH,'//div[contains(@class,"_aabd")]')
post.click()
while count!=31:
    like=driver.find_element(By.XPATH,'//span[@class="_aamw"]')
    st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
    if st=="Like":
        print("---> Post Like Successfully")
        like.click()
        
    elif st=="Unlike":
        print("You Already like Post")
    count+=1
    
    next_post=driver.find_element(By.XPATH,'//div[contains(@class,"_aaqg")]')
    next_post.click()
    time.sleep(2)

print()   
if count==31:
    print("You Liked/Unlike All 30 Posts")
        
        
```

```python
#UnLiking top 30 posts
import time
post_btn=driver.find_element(By.CLASS_NAME,'_aa_w')
post_btn.click()

time.sleep(2)
count=1

post=driver.find_element(By.XPATH,'//div[contains(@class,"_aabd")]')
post.click()
while count!=31:
    like=driver.find_element(By.XPATH,'//span[@class="_aamw"]')
    st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
    if st=="Unlike":
        print("---> Unlike Post Successfully")
        like.click()
        
    elif st=="Like":
        print("You Already Unlike Post")
    count+=1
    
    next_post=driver.find_element(By.XPATH,'//div[contains(@class,"_aaqg")]')
    next_post.click()
    time.sleep(2)
    
print()
if count==31:
    print("You Liked/Unlike All 30 Posts")
        
```

```python
import time
current_height=driver.execute_script('return document.body.scrollHeight;')

count=1
while True:
    username=driver.find_elements(By.XPATH,'//span[contains(@class,"_aap8")]/a/span')
    for i in username:
        print(i.text)
        count+=1
        if count>500:
            break
    
    if count>500:
        break
    driver.execute_script('window.scrollTo(0,arguments[0]);',current_height)
    time.sleep(2)
    
    new_height=driver.execute_script('return document.body.scrollHeight;')
    current_height=new_height

print(count)   #--->501
```

```python
try:
    story=driver.find_element(By.CLASS_NAME,'_aa8h')

    check_story=driver.find_element(By.CLASS_NAME,'_aarh')
    #Before Viewing the Story, Height of CODING NINJAS Logo is 277
    #And after Viewing the Story Height becomes 274
    
    height=check_story.get_attribute('height')
    print(height)

    if height==277:
        print("Viewing The Story")
        story.click()
    else:
        print("Story is Already Viewed")

except:
    print("No Story is Available to View")
```
