```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df_train=pd.read_csv('Downloads/training_titanic_x_y_train.csv')
df_test=pd.read_csv('Downloads/test_titanic_x_test.csv')
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.describe(include='all')
```

```python
df_test.describe(include='all')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
#Filtering and cleaning Data of Train Data

del df_train['Name']
df_train.loc[df_train['Sex']=='male','Sex']=1
df_train.loc[df_train['Sex']=='female','Sex']=2
df_train.loc[df_train['Age'].isnull(),'Age']=df_train.loc[df_train['Age'].notnull(),'Age'].mean()
del df_train['Ticket']
df_train.loc[df_train['Cabin'].isnull(),'Cabin']=0
df_train.loc[df_train['Cabin'].notnull(),'Cabin']=1
df_train.loc[df_train['Embarked']=='C','Embarked']=1
df_train.loc[df_train['Embarked']=='Q','Embarked']=2
df_train.loc[df_train['Embarked']=='S','Embarked']=3
df_train.loc[df_train['Embarked'].isnull(),'Embarked']=df_train.loc[df_train['Embarked'].notnull(),'Embarked'].median()

df_train.head()
```

```python
df_train.describe(include='all')
```

```python
data_train=df_train.values
x_train=data_train[:,:-1].astype('int')
y_train=data_train[:,-1].astype('int')
x_train.shape
```

```python
#Filtering and cleaning of Testing data

del df_test['Name']
df_test.loc[df_test['Sex']=='male','Sex']=1
df_test.loc[df_test['Sex']=='female','Sex']=2
df_test.loc[df_test['Age'].isnull(),'Age']=df_test.loc[df_test['Age'].notnull(),'Age'].mean()
del df_test['Ticket']
df_test.loc[df_test['Cabin'].isnull(),'Cabin']=0
df_test.loc[df_test['Cabin'].notnull(),'Cabin']=1
df_test.loc[df_test['Embarked']=='C','Embarked']=1
df_test.loc[df_test['Embarked']=='Q','Embarked']=2
df_test.loc[df_test['Embarked']=='S','Embarked']=3
df_test.loc[df_test['Embarked'].isnull(),'Embarked']=df_test.loc[df_test['Embarked'].notnull(),'Embarked'].median()


df_test.head()
```

```python
df_test.describe(include='all')
```

```python
data_test=df_test.values
x_test=data_test.astype('int')
x_test.shape
```

```python
algo1=LogisticRegression()
algo1.fit(x_train,y_train)
```

```python
y_train_pred=algo1.predict(x_train)
```

```python
print(abs(y_train_pred-y_train))
```

```python
l=y_train_pred-y_train
count=0
for i in l:
    if i==0:
        count+=1
        
print("Total no. of Accurate Predictions:",count)
```

```python
y_pred=algo1.predict(x_test)
print(y_pred)
```

```python
np.savetxt('predictions_LogisticReg.csv',y_pred)
```
